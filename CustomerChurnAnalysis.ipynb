```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#To open csv file
df=pd.read_csv('telco.csv')
df.head()
```

```python
#Basis information about Data
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
#Cheack null value
df.isna().sum()
```

```python
#check Duplicate values
df.duplicated().sum()
```

```python
#Deal with missing value
df['Offer'].value_counts()
```

```python
df['Offer']=df['Offer'].fillna('No offer')
```

```python
df['Internet Type'].value_counts()
```

```python
df['Internet Type']=df['Internet Type'].fillna('No Interner')
```

```python
df['Churn Category']=df['Churn Category'].fillna('no churn')
```

```python
df['Churn Reason']=df['Churn Reason'].fillna('no reasion')
```

```python
df.isna().sum()
```

```python
# saved the cleaned data 
df.to_csv('cleaned_churn_data.csv',index=False)
```

```python
# gender distribution 
sns.countplot(x=df['Gender'],color='red')
plt.title("Gender distribute")
```

```python
#churn by gender
sns.countplot(x=df['Gender'],hue=df['Churn Label'])
plt.title("Churn by the Gender")
```

```python
#contract distribute
sns.countplot(x=df['Contract'])
plt.title("Contract distribute")
```

```python
#Churn by the contract
sns.countplot(x=df['Contract'],hue=df['Churn Label'])
```

```python
#Churn by the Payment method
sns.countplot(x=df['Payment Method'],hue=df['Churn Label'])
```

```python
#Monthly charge distribution
sns.boxplot(x=df['Churn Label'],y=df['Monthly Charge'],palette='coolwarm')
```

```python
#Tenure Distribution
sns.histplot(x=df['Tenure in Months'],y=df['Churn Label'],bins=30,kde=True)
plt.title("Tenure Distribution")
```

```python
#Churn by the Sinior citizen
sns.countplot(x=df['Senior Citizen'],hue=df['Churn Label'])
plt.title("Churn by the Sinior citizen")
```

```python
#avarage charge accross payment method
sns.barplot(x=df['Monthly Charge'].mean(),y=df['Payment Method'])
plt.title("#avarage charge accross payment method")
```
